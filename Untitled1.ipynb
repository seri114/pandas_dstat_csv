```python
import pandas as pd
import numpy as np
import tempfile

pd.set_option('display.max_columns', 100)
file='dstat/dstat.csv'

def read_dstat_csv(filepath: str):
    with tempfile.TemporaryFile("w+") as ft:
        with open(filepath) as f:
            for str in f:
                ft.write(str.rstrip(",\n")+"\n")
        ft.seek(0)
        df = pd.read_csv(ft, skiprows=4, header=1)
    return df

df = read_dstat_csv(file)
df
```

```python
df2=df.loc[:,['time','used','/:used','used','free','buff','cach','cpu0 usage:usr']]
df2
```

```python
df2.describe().loc[['mean','min','max'],:]
```
